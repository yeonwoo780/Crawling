### selenium

```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.wait import WebDriverWait
options = Options()
options.headless = True
browser = webdriver.Chrome(executable_path="C:\\Users\\ad\\Downloads\\chromedriver_win32\\chromedriver.exe", options=options)
browser.get("https://finance.naver.com/sise/lastsearch2.nhn")
time.sleep(3)
tag_names = browser.find_elements_by_tag_name("table")
jusic = []
for tag in tag_names:
    jusic.append(tag.text.split("\n"))
    
jusic
```

```python
a = jusic[1][0]
a = a.split(' ')
a
```

```python
import pandas as pd
jusic_1 = jusic[1]
df_jusic_1 = pd.DataFrame(jusic_1)
df_jusic_1 = df_jusic_1[0].str.split(' ')
df_jusic_1=df_jusic_1.apply(lambda x: pd.Series(x))
df_jusic_1.reset_index(drop=True)
df_jusic_1 = df_jusic_1[df_jusic_1.index != 0].copy()
df_jusic_1
```
